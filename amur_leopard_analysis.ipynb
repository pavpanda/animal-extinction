```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = [1955, 1995, 2000, 2003, 2007, 2016, 2020]
y = [2400, 300, 25, 29, 30, 60, 84]
plt.style.use('ggplot')
plt.plot(x, y)
plt.title("Amur Leopard population vs Time")
plt.xlabel("Time (year)")
plt.ylabel("Amur Leopard population")
plt.savefig("amurwhole.png")
```

```python
plt.plot(x[2:], y[2:])
plt.xticks(range(2000, 2021, 3))
plt.title("Amur Leopard population vs Time")
plt.xlabel("Time (year)")
plt.ylabel("Amur Leopard population")
plt.savefig("amurend.png")
```
